```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Esra = pd.read_excel('output_Esra.xlsx').drop(columns=['Unnamed: 0'])
Esra
```

```python
Emy = pd.read_excel('output_Emy.xlsx').drop(columns=['Unnamed: 0'])
Emy
```

```python
Iris = pd.read_excel('output_Iris.xlsx').drop(columns=['Unnamed: 0'])
Iris
```

```python
score = pd.read_csv('ScoreATTTable.csv', sep = ',')
score
```

```python
score_low = score['0']
score_mid = score['1']
score_high = score['2']
```

```python
low_Esra = Esra.iloc[0]
mid_Esra = Esra.iloc[1]
high_Esra = Esra.iloc[2]

low_Emy = Emy.iloc[0]
mid_Emy = Emy.iloc[1]
high_Emy = Emy.iloc[2]

low_Iris = Iris.iloc[0]
mid_Iris = Iris.iloc[1]
high_Iris = Iris.iloc[2]
```

```python
x = np.arange(39)
plt.plot(x, low_Esra)
plt.plot(x, low_Emy)
plt.plot(x, low_Iris)
plt.legend(['i7-7500U CPU 2.70GHz 2.90 GHz', 'i5-6200U CPU 2.30GHz 2.40 GHz', 'i5 1.4 GHz'], loc='upper left')
plt.xlabel('Number of binary predictors')
plt.ylabel('Training time (sec)')
plt.title("A. Training time on 1/3 of the records")
plt.savefig('one-third.png')
```

```python
x = np.arange(39)
plt.plot(x, mid_Esra)
plt.plot(x, mid_Emy)
plt.plot(x, mid_Iris)
plt.legend(['i7-7500U CPU 2.70GHz 2.90 GHz', 'i5-6200U CPU 2.30GHz 2.40 GHz', 'i5 1.4 GHz'], loc='upper left')
plt.xlabel('Number of binary predictors')
plt.ylabel('Training time (sec)')
plt.title("B. Training time on 2/3 of the records")
plt.savefig('two-third.png')
```

```python
x = np.arange(39)
plt.plot(x, high_Esra)
plt.plot(x, high_Emy)
plt.plot(x, high_Iris)
plt.legend(['i7-7500U CPU 2.70GHz 2.90 GHz', 'i5-6200U CPU 2.30GHz 2.40 GHz', 'i5 1.4 GHz'], loc='upper left')
plt.xlabel('Number of binary predictors')
plt.ylabel('Training time (sec)')
plt.title("C. Training time on all records")
plt.savefig('all.png')
```

```python
x = np.arange(39)
plt.plot(x, score_low)
plt.plot(x, score_mid)
plt.plot(x, score_high)
plt.legend(['1/3 of records', '2/3 of records', 'all records'], loc='upper left')
plt.xlabel('Number of binary predictors')
plt.ylabel('Accuracy (r2)')
plt.title("D. Accuracy of the models")
plt.savefig('accuracy.png')
```
